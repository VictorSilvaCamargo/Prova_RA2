&= 0.00182670920453971

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return (x**2 - 4)*(x**3 - 8*x - 1)

x = np.linspace(-10, 10, 1000)
y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x)=(x^2−4)(x^3−8x−1)')
plt.grid()
plt.show()
```

# Questão 2 (vale 2 pontos)
Uma empresa espacial está tentando descobrir o ponto mínimo da órbita de um satélite cuja função orbital é dada por: 
$$f(x)=7\sin(x)-2\cos(x)$$

Encontre o primeiro ponto de mínimo no intervalo $-10<=x<=10$.

Para resolver esta questão você precisará de, no minimo, 4 casas decimais de precisão depois da vírgula. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a função
def f(x):
    return 7*np.sin(x) - 2*np.cos(x)

# Define o intervalo de x
x = np.linspace(-10, 10, 1000)

# Plota a função
plt.plot(x, f(x))

# Encontra os pontos críticos
minimo = 0
crit_points = []
for n in range(-10, 11):
    x1 = np.arctan(-7/2) + np.pi*n
    x2 = np.arctan(7/2) + np.pi*n
    if -10 <= x1 <= -7:
        crit_points.append(x1)
        minimo = x1

# Plota os pontos críticos em vermelho
plt.plot(crit_points, f(crit_points), 'ro')

# Mostra o gráfico
print(minimo)
plt.show()
```

# Questão 3 (vale 2 pontos)
Considerando a função $f(x,y) = \sin(xy) + \cos(x+2y)$, no intervalo $0\leq x\leq 5 \space\space e \space\space 0\leq y\leq 5$ trace, usando o plotly o gráfico desta função e marque os pontos de máximo e mínimo neste intevalo.

Lembre-se o sympy é sempre uma opção e você pode usar qualquer técnica, algoritmo, regra, ou tabela, que tenha sido usada na matéria. E que precisa de apenas uma casa decimal de precisão. Uma dica importante para a solução deste problema é a função **np.unravel_index**


```python
import plotly.graph_objs as go
import numpy as np

# Define a função f(x,y)
def f(x, y):
    return np.sin(x*y) + np.cos(x + 2*y)

# Crie uma malha de pontos (x,y)
x_vals = np.linspace(0, 5, 100)
y_vals = np.linspace(0, 5, 100)
x_mesh, y_mesh = np.meshgrid(x_vals, y_vals)

# Calcule os valores de f(x,y) para cada ponto na malha
f_mesh = f(x_mesh, y_mesh)

# Encontre os pontos críticos da função
grad_f = np.gradient(f_mesh)
zero_grad_mask = np.isclose(grad_f[0], 0) & np.isclose(grad_f[1], 0)
x_crit, y_crit = x_mesh[zero_grad_mask], y_mesh[zero_grad_mask]

# Plote o gráfico da função f(x,y) com os pontos críticos
fig = go.Figure(data=[go.Surface(z=f_mesh)])
fig.add_trace(go.Scatter3d(x=x_crit.flatten(), y=y_crit.flatten(), z=f(x_crit.flatten(), y_crit.flatten()), mode='markers', marker=dict(size=5)))
fig.show()
```
